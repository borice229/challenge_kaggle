```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
# Charger le fichier CSV dans un DataFrame
Xty = pd.read_csv("X_train.csv", sep=",")
# Convertir en datetime avec utc=True pour unifier les fuseaux horaires
Xty['DELIVERY_START'] = pd.to_datetime(Xty['DELIVERY_START'], errors='coerce', utc=True)
Xty['DELIVERY_START'] = Xty['DELIVERY_START'].dt.tz_localize(None)
```

```python
Y_dt = pd.read_csv("y_train.csv",sep=",")
Y_dt.head(2)
```

```python
# Convertir en datetime avec utc=True pour unifier les fuseaux horaires
Y_dt['DELIVERY_START'] = pd.to_datetime(Y_dt['DELIVERY_START'], errors='coerce', utc=True)

# Convertir en heure locale, puis supprimer le fuseau horaire pour garder l'heure inchangée
Y_dt['DELIVERY_START'] = Y_dt['DELIVERY_START'].dt.tz_convert('Europe/Paris').dt.tz_localize(None)
```

```python
data_merge = pd.merge(Y_dt, Xty, on="DELIVERY_START")
```

```python
Y_dt.info()
```

```python
# Renommer une colonne
data_merge.rename(columns={"DELIVERY_START": "ds","spot_id_delta":"y"}, inplace=True)
```

```python
# Renommer une colonne
Y_dt.rename(columns={"DELIVERY_START": "ds","spot_id_delta":"y"}, inplace=True)
```

```python
# Fractionner les données (80% pour entraînement, 20% pour test)
train_size = int(0.8 * len(data_merge))
train = data_merge.iloc[:train_size]  # Première partie pour l'entraînement
test = data_merge.iloc[train_size:]  # Reste pour les tests
```

```python
# Fractionner les données (80% pour entraînement, 20% pour test)
train_size = int(0.8 * len(Y_dt))
train = data_merge.iloc[:train_size]  # Première partie pour l'entraînement
test = data_merge.iloc[train_size:]  # Reste pour les tests
```

```python
train.tail(10)
```

```python
# Créer et ajuster le modèle
m = Prophet()
m.add_regressor('load_forecast')  # Exemples de variables supplémentaires
m.add_regressor('predicted_spot_price')
m.fit(train)

# Faire des prédictions
future = test[['ds', 'load_forecast', 'predicted_spot_price']]
forecast = m.predict(future)

# Évaluer les performances
y_true = test['y'].values
y_pred = forecast['yhat'].values
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}, MAPE: {mape:.2f}%")

# Visualisation
plt.figure(figsize=(12, 6))
plt.plot(test['ds'], y_true, label='Valeurs réelles', color='blue')
plt.plot(test['ds'], y_pred, label='Prédictions', color='orange')
plt.fill_between(test['ds'], 
                 forecast['yhat_lower'], 
                 forecast['yhat_upper'], 
                 color='orange', alpha=0.2, label='Intervalle de confiance')
plt.legend()
plt.title('Comparaison entre valeurs réelles et prédictions')
plt.show()
```
